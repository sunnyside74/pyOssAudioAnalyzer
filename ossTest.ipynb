```python
# Plot loaded wave file

plot_time = np.linspace(0., time, wav_data.shape[0])

plt.plot(plot_time, wav_data[:, 0], label="Left channel")
plt.plot(plot_time, wav_data[:, 1], label="Right channel")
plt.legend()
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")
plt.show()
```
